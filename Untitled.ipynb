Ex 1.) Looking into the Lorenz-63 system with an error term given by the tent map on [0,1]. At first we will not account for the error  term $g_i^n$ generated by the tent map.

```python
sigma = 10.0
rho = 28.0
beta = 8.0/3.0
dt = 0.01
x = np.array([[5,0,1]])
inter = 4000
t = 0

def f(x):
    dxdt = np.array([sigma*(x[1] - x[0]), x[0]*(rho - x[2]) - x[1], x[0]*x[1] - beta*x[2]])
    return dxdt
    
for k in range(1,inter + 1):
    step = x[-1,:] + dt*(f(x[-1,:]))
    x = np.append(x, np.array([step]), axis = 0)
    
fig = plt.figure(figsize = (20,20))
ax = fig.gca(projection='3d')
ax.plot(x[:,0], x[:,1], x[:,2], label='Lorentz-63')
ax.legend()

plt.show()
```

Reducing now the amount of points saved to those at times $\Delta t_{out}$ and accounting for the error term.

```python
sigma = 10.0
rho = 28.0
beta = 8.0/3.0
dt = 0.01
x = np.array([[0,0,0]])
step = np.array([0,0,0])
inter = 4000
t = 0

def f(x):
    dxdt = np.array([sigma*(x[1] - x[0]), x[0]*(rho - x[2]) - x[1], x[0]*x[1] - beta*x[2]])
    return dxdt

def tent(t):
    IndInc = (t < 1.0/2.0) & (t < 0)
    IndDec = (t <= 1.0) & (t < 1.0/2.0)
    return 1.9*t*IndInc + ((-1.9)*t + 1.9)*IndDec
 
    
for k in range(1,5*inter + 1):
    g0 = tent(step[0])# andere fehler
    g1 = tent(step[1])
    g2 = tent(step[2])
    step = step + dt*(f(step) + np.array([g0,g1,g2])) 
    if k%5 == 0:
        x = np.append(x, np.array([step]), axis = 0)


np.savetxt('Observed points', x)

        
#fig = plt.figure()
#ax = fig.gca(projection='3d')
#ax.plot(x[:,0], x[:,1], x[:,2], label='Lorenz-63')
plt.plot(range(100), x[:100,0])
plt.xlabel('time t')
plt.ylabel('x-coordinate')
plt.legend()

plt.show()
```

Ex. 3) Forecasting for intervals $\Delta t_{out}=0.05$

```python
y_obs = np.loadtxt('Observed points')[:,0]
d = np.shape(y_obs)[0]

y_pred = np.array([y_obs[:2]])

for i in range(1,d-1):
    pred = 2*y_obs[i] - y_obs[i-1]
    y_pred = np.append(y_pred, pred)
    
RMSE = np.sqrt(np.sum(1/d*np.absolute(y_obs - y_pred)**2))
print(RMSE)


plt.plot(range(100), y_obs[:100], 'bo-')
plt.plot(range(100), y_pred[:100], 'rx-')
plt.xlabel('time t')
plt.ylabel('x-coordinate')
plt.legend(('y_obs', 'y_predict'))

plt.show()
```

Forecasting for intervals $\Delta t_{out}=0.1$

```python
y_obs = np.loadtxt('Observed points')[:,0]
d = np.shape(y_obs)[0]

y_pred = np.array([y_obs[:2]])

for i in np.arange(2,d-1,2):
    pred = 2*y_obs[i] - y_obs[i-2]
    y_pred = np.append(y_pred, pred)
    
RMSE = np.sqrt(np.sum(1/d*np.absolute(y_obs[np.arange(0,d,2)] - y_pred)**2))
print(RMSE)



plt.plot(range(100), y_obs[:100], 'bo-')
plt.plot(np.arange(0,100,2), y_pred[:50], 'rx-')
plt.xlabel('time t')
plt.ylabel('x-coordinate')
plt.legend(('y_obs', 'y_predict'))

plt.show()
```
